```python
import pandas
from tokenizers import Tokenizer,models,pre_tokenizers,trainers
import tokenizers
from tqdm import tqdm
```

```python
VOCAB_SIZE = 10000
```

```python
path_to_train_ds = "path to train dataset"
train_ds = pandas.read_parquet(path_to_train_ds)

train_ds_array =[]

for i in train_ds.iloc:
    train_ds_array.append(i.text)
    
```

```python
tokenizer = Tokenizer(models.BPE(unk_token="[UNK]"))
tokenizer.pre_tokenizer= pre_tokenizers.Whitespace()


trainer = trainers.BpeTrainer(vocab_size = VOCAB_SIZE,min_frequency = 4,show_progress = True,special_tokens=["[PAD]","[END]","[START]","[UNK]"])
tokenizer.train_from_iterator(train_ds_array,trainer)

tokenizer.save("SLM_tokenizer.json")
```
